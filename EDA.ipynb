```python
import pandas as pd

df = pd.read_csv("D:/.Semester 7/FYP/FYP docx/data.csv", encoding="ISO-8859-1")
```

```python
#country names
total_countries = df['Country'].unique()
total_countries
```

```python
#countries with sales
country_order_count = df.groupby(['Country']).size().reset_index()
country_order_count.sort_values(0, ascending=False)
```

```python
#total unique customers
total_customers = df['CustomerID'].unique()
total_customers

unique_customerIDs = pd.DataFrame({'unique CustomerIDs': total_customers[:]})
unique_customerIDs
#print('total number of customers = ', total_customers.shape[0])
```

```python
#least recent and most recent order dates
#time period of data
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
least_recent_date = df['InvoiceDate'].min()
recent_date = df['InvoiceDate'].max()
print('most recent order invoice date  = ', recent_date)
print('least recent order invoice date  = ', least_recent_date)
print('total records of = ', recent_date-least_recent_date)
```

```python
#total unique products

total_products = df['Description'].unique()
total_products

unique_descriptions = pd.DataFrame({'unique products': total_products[:]})
unique_descriptions
```

```python
#creating new .csv file 
new_df = df.groupby(['InvoiceNo', 'InvoiceDate', 'CustomerID', 'Country'])['UnitPrice'].sum()
new_dff = new_df.to_frame()

export_csv = new_dff.to_csv (r'D:\.Semester 7\FYP\FYP docx\data_updated.csv', header=True)
new_dff
```
